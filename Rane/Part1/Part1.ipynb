Problem 1 :
In this problem, we take two vectors - sales1 and sales2, where sales1 is defined and sales2 is given random values. 'par' can be used to set graphical parameters, in this case, we set the background color to "cornsilk". We first use the plot function to mark x and y axes, specify sales1 line color. We then add a title for the graph and then set color and type of sales2 using lines(). We then add a legend. On every execution, we see a difference in the graph as sales2 has different values on every iteration.

```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
sales2<-rpois(12,34)
par(bg="cornsilk")

plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales" )
title(main="Sales by Month")

lines(sales2, type="o", pch=22, lty=2, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
```

Problem 3 :
In this problem, we plot a boxplot using data. We have also created a boxplot for the dataset "BOD", where objects 'dist' and 'time' are masked in the boxplot. In boxplots, we get values such as the upper and lower extremes, the upper and lower quartile and median. Also, we find the quartiles by using the function quantile().


```r
fn<-boxplot(sales,col=c("red","green"))$stats

text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0, cex=.7)
quantile(fn, c(0.25,0.5,0.75))
grid(nx=NA, ny=NULL)
```

```r

fn<-boxplot(BOD,col=c("yellow","blue"))$stats

text(1.45, fn[1,2], paste("Median =", fn[1,2]), adj=0, cex=.7)
text(0.45, fn[2,1],paste("Median =", fn[2,1]), adj=0, cex=.7)
quantile(fn, c(0.25,0.5,0.75))
grid(nx=NA, ny=NULL)
```

Problem 4: 
In this problem, we import two csv files and plot lines similar to problem 1. We can see the price drop of "Apple" indicated in blue.
We also plot histograms for both datasets using the 'hist' function.
Both datasets show a great rise between 175 and 180, though the frequency for Apple drops right after 175.

```r
a<-read.csv("AAPL.csv",header = TRUE)
b<-read.csv("FB.csv",header = TRUE)
par(bg="cornsilk")
plot(a$Adj.Close, col="blue", type="o", ylim=c(150,200), xlab="Days", ylab="Price" )
lines(b$Adj.Close, type="o", pch=22, lty=2, col="red")
legend("topright", inset=.05, c("Apple","Facebook"), fill=c("blue","red"), horiz=TRUE)

hist(a$Adj.Close, col=rainbow(8))
hist(b$Adj.Close, col=rainbow(8))
```
